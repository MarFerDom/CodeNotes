### Displaying audio

```python
#| export
import numpy as np
from IPython.display import Audio
```

```python
t = np.linspace(0,5,44100*5)
sine = np.sin(2*np.pi*520*t)
audio = Audio(sine, rate=44100)

display(audio)
```

### Displaying Progress Bar

```python
#| export
from IPython.display import ProgressBar
import time
```

HTML style

```python
pb = ProgressBar(12)
for _ in pb: time.sleep(0.5)
```

Old school

```python
for i in range(11):
    pb.progress = i
    print('\r', pb, sep='', end='')
    time.sleep(0.5)
```

Changing `ProgressBar` properties

```python
pb2 = ProgressBar(35)
pb2.html_width = '70ex'
pb2.text_width = 50
pb2.display()
```

```python
for i in range(11):
    pb2.progress = i
    print('\r', pb2, sep='', end='')
    time.sleep(0.5)
```

## PyAutoGUI

```python
#| export
#|eval: false
import pyautogui as pa
import random
```

```python
#|eval: false
pa.FAILSAFE = True
w,h = pa.size()
random.seed(42)

def moveWithin():
    pa.moveTo(random.randint(0,w),random.randint(0,h), duration=2)

while True:
    try:
        moveWithin()
    except pa.FailSafeException:
        print(f"Stopping with the randomness 'cause: {pa.FailSafeException}")
        break
print(pa.position())
```

## Redlines

Find differences between two texts

```python
#| export
from IPython.display import display, Markdown
from redlines import Redlines
```

```python
t1 = 'She sells seashells in Seychelles'
t2 = 'She shells seysells in Seachelles'
diff = Redlines(t2, t1)
display(Markdown(diff.output_markdown))
```
